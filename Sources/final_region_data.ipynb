```python
"""
관광지 데이터

1. 주소 가져오지 못한 부분은 직접 검색해서 넣어줌
2. 한 장소의 부속장소를 합침 (한림공원, 한림공원 왕벚꽃동산 -> 한림공원)
3. 폐쇄된 장소, 휴관중인 장소는 삭제함
4. 관광지 성격과 맞지 않는 장소는 삭제함(편의점, 주차장, 식당 등)
5. 산봉우리는 제외함(00봉)

"""
```

```python
import folium
from folium.plugins import MiniMap

# 지도를 표출하는 함수
def make_map(dfs):
    # 기준지 - 제주국제공항
    m = folium.Map(location=[33.5101562, 126.4861157], zoom_start=12)
    # 미니맵 추가
    minimap = MiniMap()
    m.add_child(minimap)
    # 마커 추가
    for i in range(len(dfs)):
        folium.Marker([dfs['y'][i], dfs['x'][i]],
                      tooltip=dfs['stores'][i],
                      popup=dfs['place_url'][i],).add_to(m)
    return m 
```

```python
make_map(tour_region_sorted) #관광지
```

```python
make_map(tour_cafe_souvenir_data) #카페/기념품
```

```python
make_map(tour_restaurant_sorted) #식당
```
