```python
"""
관광지 데이터

1. 주소 가져오지 못한 부분은 직접 검색해서 넣어줌
2. 한 장소의 부속장소를 합침 (한림공원, 한림공원 왕벚꽃동산 -> 한림공원)
3. 폐쇄된 장소, 휴관중인 장소는 삭제함
4. 관광지 성격과 맞지 않는 장소는 삭제함(편의점, 주차장, 식당 등)
5. 산봉우리는 제외함(00봉)

"""
```

```python
import pandas as pd

tour_region_data = pd.read_csv("tour_region.csv", encoding="cp949")
tour_region_data.head()
```

```python
tour_region_sorted = tour_region_data.sort_values(by=['road_address'])
display(tour_region_sorted.head())
```

```python
tour_region_sorted.to_csv("tour_region.csv", index=False, encoding="cp949")
```
